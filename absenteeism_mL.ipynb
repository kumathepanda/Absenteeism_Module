```python
from sklearn.base import BaseEstimator, TransformerMixin
class CustomScaler(BaseEstimator,TransformerMixin):
    def __init__(self,columns , copy=True,with_mean=True,with_std=True):
        self.columns = columns
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.mean_ = None
        self.var_ = None
        
    def fit(self,X,y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_= np.mean(X[self.columns])
        self.var_= np.var(X[self.columns])
        return self
    
    def transform(self,X,y=None,copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]),columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_scaled,X_not_scaled],axis=1)[init_col_order]
        
```

```python
columns_to_scale = [cols for cols in unscaled_inputs.columns if cols not in ["Education","reason_type_1","reason_type_2","reason_type_3","reason_type_4"]]
```

```python
columns_to_scale
```

```python
absenteesim_scaler = CustomScaler(columns_to_scale)
```

```python
absenteesim_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteesim_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
reg = LogisticRegression()
X_train,X_test,y_train,y_test=train_test_split(scaled_inputs,targets,test_size=0.25,random_state=42,shuffle=True)
```

```python
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
accuracy_score(y_test,reg.predict(X_test))
```

```python
intercept = reg.intercept_
coefs = reg.coef_
summary_table = pd.DataFrame(columns=["feature names"],data=feature_names)
summary_table["Coefficients"] = np.transpose(coefs)
summary_table.index = summary_table.index+1
summary_table.loc[0] = ["intercept",intercept[0]]
summary_table = summary_table.sort_index()
summary_table["odds"] = np.exp(summary_table["Coefficients"])
summary_table.sort_values(by="odds",ascending=False)
```
